So far we have basicly look at maximum likelihoods for linear functions and solved them like we only know calculus (grad) and not linear algebra (qr decomposition). Lets try and do something nonlinear and with layers forcing use to use grad. This particular example comes from pg. 231 of Bishiop. 

Our goal will be to find the gradiant of our error function in other words, find

 $$ \nabla E(w)$$.
 
 We will define the $i$th erorr function before as one half the sum of the diffrence of squares which has a derivative,
 
 $$ E(w) = \sum \frac{(\hat{y}_{i}-y_{i})^{2}}{2}$$ 
 
Now we must use the chain rule to find it's derivative. First we must examine the out put $\hat{y}_{k}$ 

$$ a_{j}^{(1)} = \sum^{D}_{i=0} w^{(1)}_{ij}x_{i}, \quad z_{i} = h(a_{i}) $$


So we see that $E(w)$ depends on $w,a,z,x$

So, the derivative with respect to $w$ for the last layer 

$$ \frac{\partial E(w)}{\partial w^{(2)}} = \frac{ \partial E}{\partial \hat{y}} \frac{ \partial \hat{y}}{\partial w} = (\hat{y} - y)z $$

Now with respect to the weights from the first layer,

$$\frac{\partial E}{\partial w^{(1)}} = \frac{\partial E}{\partial \hat{y}} \frac{\partial \hat{y}}{\partial a} \frac{\partial a}{\partial w^{(1)}} = (\hat{y} - y)(wh^{\prime}(a))(x)$$


Now we will introduce some notation from Bishop $\delta_{i} = \frac{\partial E}{\partial a_{j}}$ this will clean up our equations.

So we have found a way to find the $w$'s that will minimize the error.

```julia
sigmod(x) = 1/(1+exp(-x))
dsig(x) = exp.(x) ./ (1 .+ exp.(x)).^2
softmax(x) = exp.(x) ./ sum(exp.(x))
```

```julia
#The quadratic fuction is a good example of nonlinearity 
X = -10:10
Y = X.^2
normY = Y ./ 100
 = .2

w1 = rand(21,21) .- 0.5
b1 = rand(21) .- 0.5
w2 = rand(21,21) .- 0.5
b2 = rand(21) .- 0.5
global z2


for i in range(1,stop = 200)

    # forward prop
    a1 = w1 * X + b1
    z1 = sigmod.(a1)
    a2 = w2 * z1 + b2
    z2 = sigmod.(a2)

    #backward prop (find grad)
     = z2 - normY
    W2 =   * transpose(z1) 
    b2 =  sum()
     = transpose(w2) *  .* dsig(a1)
    W1 =   * transpose(X)
    b1 =  sum()

    #update params y1 = y0 - a * y
    w1 = w1 .-  * W1
    b1 = b1 .-  * b1
    w2 = w2 .-  * W2
    b2 = b2 .-  * b2
    
    print(sum(abs.(z2 .- normY)),"\n")
    
end

```
